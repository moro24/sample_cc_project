## Preprocessing / Cleaning Dataset 
- drop the Unnamed column
- Replace special characters
- Change the column types
- missing values 

```python
## deleting col
del df['Unnamed: 0']
```

```python
## 
df['KM'] = df['KM'].replace('??', np.nan) 
```

```python
## 
df['HP'] = df['HP'].replace('????', np.nan) 
```

```python
##
df['Doors'] = df['Doors'].replace('three', 3)
df['Doors'] = df['Doors'].replace('four', 4) 
df['Doors'] = df['Doors'].replace('five', 5) 
```

```python
##
df['KM'] = df['KM'].astype('float64') 
```

```python
## 
df['FuelType'] = df['FuelType'].astype('category')
```

```python
## 
df['HP']  = df['HP'].astype('float64')
```

```python
## 
df['MetColor'] = df['MetColor'].astype('category')
```

```python
df['Automatic'] = df['Automatic'].astype('category')
```

```python
df['Doors'] = df['Doors'].astype('category')
```

```python
df['CC'] = df['CC'].astype('category')
```

```python
## 
df['KM'] = df['KM'].fillna(df['KM'].mean())
```

```python
## 
df['HP'] = df['HP'].fillna(df['HP'].mean())
```

```python
## 
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['MetColor'].value_counts()
```

```python
df['MetColor'] = df['MetColor'].fillna(df['MetColor'].value_counts().index[0])
```

```python
df['FuelType'].value_counts()
```

```python
df['FuelType'] = df['FuelType'].fillna(df['FuelType'].value_counts().index[0])
```

```python
df.isnull().sum()
```

```python
## 
df['KM'] = df['KM'].astype('int64')
```

```python
## 
df['Age'] = df['Age'].astype('int64')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## select only numerical datatype
df_numerical = df.select_dtypes(include=['int', 'float'])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_numerical.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
## lets plot the distribution of the target var price 
sns.histplot(df['Price'], kde=True, bins=30, color="blue")
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Age vs Price
plt.subplot(1, 2, 1)
sns.scatterplot(x='Age', y='Price', data=df, color='green')
plt.title("Age vs Price")

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# KM vs Price
plt.subplot(1, 2, 2)
sns.scatterplot(x='KM', y='Price', data=df, color='red')
plt.title("KM vs Price")

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# FuelType
plt.subplot(1, 2, 1)
sns.boxplot(x='FuelType', y='Price', data=df)
plt.title("FuelType vs Price")
```
